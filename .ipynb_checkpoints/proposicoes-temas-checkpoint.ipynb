### Create themes categories

```python
temas_tipo  = pd.read_csv('Temas_Macros.csv', sep=';', encoding='latin-1', index_col='B')
temas_tipo = temas_tipo[['C']]
temas_tipo.head()
```

```python
# Treat and separate thems

df1 = df[['idproposicao', 'tema']].set_index('idproposicao', drop=True)

d = df1['tema'].to_dict()
d = {k: v.split(';') for k, v in d.items()}
d = {k: list(map(lambda x: x.replace('(utilizado até a legislatura 53)', '').strip(), v)) for k, v in d.items()}


# Categorize themes by temas_tipo
df['tema_tipo'] = ''

for idx, temas in d.items():
    
    tipos = list(set([temas_tipo.loc[[i]]['C'].values[0] for i in temas]))
        
    df.loc[df['idproposicao'] == idx, 'tema_tipo'] = tipos[0] 
    
    if len(tipos)>1:
        for tipo in tipos[1:]:
            
            row = deepcopy(df[df['idproposicao'] == idx])
            row['tema_tipo'] = tipo
            df = df.append(row,ignore_index=True)
    else:
        pass
```

### Check sample distribution

```python
from collections import Counter
import matplotlib.pyplot as plt

def plot_pie(y):
    
    #count = y.sum(axis=0)
    #target_stats = count.to_dict()
    target_stats = Counter(y)
    
    labels = list(target_stats.keys())
    sizes = list(target_stats.values())
    explode = tuple([0.1] * len(target_stats))

    fig, ax = plt.subplots()
    ax.pie(sizes, explode=explode, labels=labels, shadow=True,
           autopct='%1.1f%%')
    ax.axis('equal')
```

```python
from sklearn.feature_selection import chi2
import numpy as np

N = 2
for tema_tipo, category_id in sorted(category_to_id.items()):
      
    features_chi2 = chi2(features, labels == category_id)
    indices = np.argsort(features_chi2[0])
    feature_names = np.array(tfidf.get_feature_names())[indices]
    
    unigrams = [v for v in feature_names if len(v.split(' ')) == 1]
    bigrams = [v for v in feature_names if len(v.split(' ')) == 2]
    
    print("# '{}':".format(tema_tipo))
    print("  . Most correlated unigrams:\n       . {}".format('\n       . '.join(unigrams[-N:])))
    print("  . Most correlated bigrams:\n       . {}".format('\n       . '.join(bigrams[-N:])))
```

## First approach: LinearSVC with probability relevance

---

**Problems:** 
- We don't use the PT stopwordsin the analysis, since `TfidfTransformer` doesn't support styled vocabulary
- It seems that the SVC classification choice it's not the one with most probability for a class.

**Advanteges:** 
- The model has 70% of accuracy in test dataset, but it generalizes well, having only 17% of classification error for all the labeled data.

```python
# Vectorizing and applying TFIDF to text

count_vect = text.CountVectorizer()
tfidf_transformer = text.TfidfTransformer()

X_train_counts = count_vect.fit_transform(X_train)
X_train_tfidf = tfidf_transformer.fit_transform(X_train_counts)
#X_train_tfidf = tfidf.fit_transform(X_train)

X_test_counts = count_vect.transform(X_test)
X_test_tfidf = tfidf_transformer.transform(X_test_counts)
#X_test_tfidf = tfidf.transform(X_test)
```

### SVC Model

```python
from sklearn.svm import LinearSVC
from sklearn.calibration import CalibratedClassifierCV


# Dá primeira rodada do TPOT, tivemos...
svm = LinearSVC(C=0.4, 
                dual=True, 
                loss='hinge', 
                penalty='l2', 
                tol=1e-05)

clf = CalibratedClassifierCV(svm) 

clf.fit(X_train_tfidf, y_train)
clf.score(X_test_tfidf, y_test)
```

```python
import itertools

lista_temas = df2.groupby(['idproposicao'])['tema_tipo'].apply(list).reset_index()

df2 = df2[['idproposicao', 'indexacao', 'tema']].drop_duplicates().merge(lista_temas, on='idproposicao')
df2.head(3)
```

```python
df3 = deepcopy(df2)
df3 = df3.set_index('idproposicao', drop=True)
df3.head()
```

```python
def category_columns_to_list(row):
    
    temas = set(row.tolist())
    
    if ' ' in temas:
        temas.remove(' ')
    
    # we put non classified in Social because is the popular category
    if len(temas) == 0:
        temas = set(['Social'])
    # TEST IF IT LOWS ERROR DOWN 
    
    return list(temas)

def choose_categories_by_prob(row, threshold):
    
    for idx in row.index:

        if row[idx] >= threshold:
            row[idx] = id_to_category[idx]
        else:
            row[idx] = ' '
    return row
```

```python
def test_thresholds_for_prob(values, prob, df3):
    
    #duplas = []
    #erro = []
    
    for i in values:

        prob_n = prob.apply(lambda x: choose_categories_by_prob(x, i), axis=1)
        prob_n = prob_n.apply(category_columns_to_list, axis=1)

        column = 'tema_tipo_est_{}'.format(round(i, 3)) #round(i, 2)
        
        if column in df3.columns:
            pass

        else:
            df3 = pd.concat([df3, prob_n], axis=1, sort=False, join='outer')
            df3 = df3.rename(columns={0: column})
        #df3 = df3.merge(prob_n, on='idproposicao', how='left')

        #erro.append(tuple([round(i, 2), len(df3[df3[column] != df3['tema_tipo']])]))
        #duplas.append(tuple([round(i, 2), len(df3[df3[column].str.len() > 2])]))
    
    return df3#, erro, duplas
```

```python
thresholds = np.linspace(0, 1, 11) # for values above 0.5, we are necessarily picking 1 or 0 categories

#df3, erro, duplas = test_thresholds_for_prob(thresholds, df3)
df3 = test_thresholds_for_prob(thresholds, prob, df3)
```
